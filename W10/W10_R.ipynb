```r
library(readr)
library(dplyr)
library(tidyverse)
library(haven)
library(margins)
require(sandwich)
```

```r
data <- read_dta(file = "~/Data/Dropbox/Working_Directory/Tex_MD/Teaching/tutorial/6002/W10/mroz9.dta")
```

```r
mod_summaries <- list()
ame_summaries <- list()
```

```r
X_vari <- c('nwifeinc', 'educ', 'exper', 'expersq', 'age', 'kidslt6')
Y_vari <- c('inlf')
```

# Probit

```r
#run the Probit regression
reg1 <- glm(get(Y_vari) ~ ., family = binomial(link = 'probit'), 
         data[ , c(Y_vari, X_vari)])

#compute the robust standard error
reg_cov<- vcovHC(reg1, type = "HC0")
std_err <- sqrt(diag(reg_cov))
q_val <- qnorm(0.975)

mod_summaries <- cbind(
Estimate = round(coef(reg1), digits = 4), 
"Robust SE" = round(std_err, digits = 4), 
z = round( (coef(reg1)/std_err), digits = 4), 
"Pr(>|z|) "= round( 2 * pnorm(abs(coef(reg1)/std_err), lower.tail = FALSE), 
                    digits = 4), 
LL = round(coef(reg1) - q_val  * std_err, digits = 4), 
UL = round(coef(reg1) + q_val  * std_err, digits = 4)
)

#compute the average marginal effect under robust standard error
ame_summaries <- margins_summary(reg1, vce = c("bootstrap"), vcov = vcovHC(reg1))
```

```r
summary(reg1)
```

```r
mod_summaries
```

```r
ame_summaries
```
