```stata
qui ac     prcfat, name(g1_prcfat, replace)
qui pac    prcfat, name(g2_prcfat, replace)
qui tsline prcfat, name(g3_prcfat, replace)

graph combine g1_prcfat g2_prcfat g3_prcfat, col(1) name(combined, replace)
```

```stata
qui ac     unem, name(g1_unem, replace)
qui pac    unem, name(g2_unem, replace)
qui tsline unem, name(g3_unem, replace)

graph combine g1_unem g2_unem g3_unem, col(1) name(combined, replace)
```
